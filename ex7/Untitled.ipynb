```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
mu = np.zeros(2)
cov = np.eye(2)

samples = np.random.multivariate_normal(mu,cov,1000)
```

```python
plt.plot(samples[:,0],samples[:,1],".")
```

```python
meanData = np.apply_along_axis(np.mean,0,samples)
covData = np.cov(np.transpose(samples))
```

```python
from scipy.stats import multivariate_normal
```

```python
prob_samples = multivariate_normal.pdf(samples,meanData,covData)
threshold = 0.01*(np.max(prob_samples))
np.max(prob_samples)
anom = samples[prob_samples<threshold]
plt.plot(anom[:,0],anom[:,1],'+')
plt.plot(samples[:,0],samples[:,1],".")
```
